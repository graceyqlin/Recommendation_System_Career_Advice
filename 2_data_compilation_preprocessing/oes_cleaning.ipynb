```python
import pandas as pd
import numpy as np
```

```python
oes = pd.read_csv('../../data/oes_report.csv')
oes.head()
```

```python
oes.dtypes
```

```python
for v in [x for x in oes.columns if 'Hourly' in x]:
    oes[v] = oes[v].str.replace('-','NaN').astype(float)
oes['occupation'] = oes['Occupation (SOC code)'].apply(lambda x: x.split('(')[0])
oes['soc'] = oes['Occupation (SOC code)'].apply(lambda x: x.split('(')[1].replace(')', ''))
oes.soc = oes.soc.apply(lambda x: x[:2]+'-'+x[2:])
oes.head()
```

```python
oes.drop('Occupation (SOC code)', axis = 1, inplace = True)
oes.rename(columns = {
    'Employment(1)': 'employment',
    'Hourly 10th percentile wage': 'hourly_10',
    'Hourly 25th percentile wage': 'hourly_25',
    'Hourly median wage': 'hourly_med',
    'Hourly 75th percentile wage': 'hourly_75',
    'Hourly 90th percentile wage': 'hourly_90'
}, inplace = True)
oes.set_index('soc', inplace = True)
oes.head()
```

```python
proj = pd.read_csv('../../data/employment_projections.csv', thousands = ',')
proj.head()
```

```python
proj['soc'] = proj['SOC Code'].apply(lambda x: x.replace('"','').replace('=',''))
proj.drop(['Occupation Title','SOC Code','Occupational openings, 2016-2026 annual average (thousands)',
           '2018 median annual wage','Education Code','Workex Code','trCode'], 
          axis = 1, 
          inplace = True)
proj.rename(columns = {
    'Employment 2016 (thousands)': 'empl_2016',
    'Employment 2026 (thousands)': 'empl_2026',
    'Employment change, 2016-2026 (thousands)': 'empl_chng',
    'Employment change, 2016-2026 (percent)': 'empl_chng_pct',
    'Typical entry-level education': 'entry_edu',
    'Work experience in a related occupation': 'work_exp',
    'Typical on-the-job training': 'otj_trainig'
}, inplace = True)
proj.set_index('soc', inplace = True)
proj.head()
```

```python
final = oes.join(proj)
final.describe()
```

```python
final.to_csv('../../data/oes_clean.csv')
# final.T.to_json('../../data/oes_clean.json')
```
