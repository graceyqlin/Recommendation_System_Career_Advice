```python
import docx2txt
import os
import glob
import pandas as pd
import re
from docx_resume_to_df import convert_to_df
from nltk import word_tokenize
```

```python
# change directory to dataset location
os.chdir('/Users/lucyxie/Documents/MIDS/Capstone/data-science-for-good-careervillage/data/Resume&Job_Description/Original_Resumes/')
```

```python
# get all the file names with .docx extension
docx_list = []
for file in glob.glob('**/*.docx', recursive=True):
    docx_list.append(file)
```

```python
# iterate through list of docx files and convert each to a dataframe
resume_df = pd.DataFrame()
for docx in docx_list:
    docx_body = docx2txt.process(docx)
    test_df = convert_to_df(docx, docx_body)
    # append to master resume dataframe
    resume_df = resume_df.append(test_df)
```
